```python
df=pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2021-01.parquet')
df
```

```python
df=df[df.trip_type==2]
df
```

```python
# Transformer les colonne date en datetime
df.lpep_pickup_datetime=pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime=pd.to_datetime(df.lpep_dropoff_datetime)
```

```python
df['duration']=df.lpep_dropoff_datetime-df.lpep_pickup_datetime
```

```python
df
```

```python
td=df.duration.iloc[0]
td
```

```python
td.total_seconds()/60 pour mettre le temps en minutes
```

```python
# transformation de la colonnes duration en times 

df.duration=df.duration.apply(lambda td: td.total_seconds()/60)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.displot(df.duration)
```

```python
df.duration.describe(percentiles=[0.95,0.98,0.99])
```

```python
df=df[(df.duration>=1) & (df.duration) <=60 ]
df
```

```python
categorical=['PULocationID','DOLocationID']
numerical=['trip_distance']
```

```python
df[categorical]=df[categorical].astype('str')
```

```python
from sklearn.feature_extraction import DictVectorizer
dv=DictVectorizer()
```

```python
train_dicts=df[categorical+numerical].to_dict(orient='records')
```

```python
X_train=dv.fit_transform(train_dicts)
```

```python
X_train
```

```python
target='duration'
y_train=df[target].values
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_train)
```

```python
sns.distplot(y_pred,label='Prediction')
sns.distplot(y_train,label='Actual')

plt.legend()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train,y_pred,squared=False)
```

```python
from sklearn.linear_model import Lasso
```

```python
lr=Lasso(0.001)
lr.fit(X_train,y_train)
mean_squared_error(y_train,y_pred,squared=False)
```

```python
import pickle

# Exemple : ton modèle est stocké dans une variable appelée 
with open('modele.pkl', 'wb') as f:
    pickle.dump((dv, lr), f)
```
